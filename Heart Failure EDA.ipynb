### What is the distribution of age among heart failure patients in the dataset

```python
heart['age'].value_counts().nlargest(10)
```

```python
sns.histplot(heart['age'], kde = True)
sns.kdeplot(heart['age'], color='red', linewidth=2)
plt.show()
```

- "The dataset encompasses individuals aged between 40 and 95 years, with the majority falling within the 50 to 75 age range."

### How does the death rate vary with age?

```python
plt.subplot(211)
sns.scatterplot(x='age', y='DEATH_EVENT', hue='DEATH_EVENT', data=heart)
plt.title('Death Event vs. Age')
plt.xlabel('Age')
plt.ylabel('Death Event')
plt.legend(labels=['Not Died', 'Died'])
plt.yticks([0, 1])  # Set y-axis ticks to only show 0 and 1
plt.show()

plt.subplot(212)
sns.lineplot(x='age', y='DEATH_EVENT', data=heart)
plt.title('Death Event vs. Age')
plt.xlabel('Age')
plt.ylabel('Death Event')
plt.show()
```

```python
X = heart[['age','DEATH_EVENT']]
CORR = X.corr()
CORR
```

- "Although there is a positive correlation between age and death, the relationship between these two features is not statistically significant."

### What is the percentage of male and female patients in the dataset?

```python
a = heart['sex'].value_counts()
print(a)
plt.pie(a,labels=['Male','Female'],autopct= '%.2f%%')
plt.title('Gender Distribution')
plt.show()
```

- "In the dataset, the majority of the patient population is male, comprising 64.88% of the total, while female patients    constitute 35.12%."

### How does the platelet count vary among different age groups?

```python
mean_platelets = heart['platelets'].mean()
print(f'Mean Platelet Count: {mean_platelets}')
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='platelets', data=heart,errorbar=('ci', 95))
plt.axhline(y=mean_platelets, color='red', linestyle='--', label='Mean Platelet Levels')
plt.title('Platelet Levels vs. Age')
plt.xlabel('Age')
plt.ylabel('Platelet Level')
plt.legend()
plt.show()
```

```python
x = heart['platelets'].mean()
y = heart[heart['platelets']<x]
y
```

- "In the dataset, patients with platelet counts lower than the average are slightly more numerous than those with counts higher than the mean. Additionally, there is no significant variation in platelet counts across different age groups."

### Is there a correlation between creatinine and sodium levels in the blood? 

```python
y = heart[['serum_creatinine','serum_sodium']]
corr = y.corr()
sns.heatmap(corr,annot=True)
plt.title('Creatinine vs Sodium')
```

- "There exists a weak negative correlation between creatinine and sodium levels in the blood."







### How does the prevalence of high blood pressure differ between male and female patients?

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x='sex', hue='high_blood_pressure',data= heart)
plt.title('Prevalence of High Blood Pressure by Sex')
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
plt.legend(title='High Blood Pressure', labels=['No', 'Yes'])

for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 8), textcoords='offset points')
plt.show()
```

- The dataset comprises 194 male patients and 105 female patients.
- Among male patients, 133 (68.56%) do not have high blood pressure, while 61 (31.44%) have high blood pressure.
- For female patients, 61 (58.10%) do not have high blood pressure, whereas 44 (41.90%) have high blood pressure.

### What is the relation between smoking habit and the occurrence of heart failure?

```python
ax=sns.countplot(x='smoking', hue='DEATH_EVENT',data= heart)
plt.title('Occurrence of Heart Failure due to Smoking')
plt.xlabel('Smoking (0: Non Smoker, 1: Smoker)')
plt.ylabel('Count')
plt.legend(title='Death', labels=['Not Died', 'Died'])

for i in ax.patches:
    ax.annotate(format(i.get_height(), '.0f'), (i.get_x() + i.get_width() / 2, i.get_height()),
            ha='center', va='center', xytext=(0, 6), textcoords='offset points')
plt.show()
```

- Among the individuals who did not smoke, a total of 137 (69.23%) survived, while 61 (30.77%) died.
- Conversely, among smokers, 66 (68.75%) survived, while 30 (31.25%) died.

### Are there any noticeable patterns in the distribution of death events across different age groups?

```python
sns.boxplot(x='DEATH_EVENT',y='age',data=heart)
plt.title('Distribution of Death Events Across Different Age Groups')
plt.xlabel('Death Event (0: Not Died, 1: Died)')
plt.ylabel('Age')
plt.show()
```

```python
heart.groupby('DEATH_EVENT')['age'].median()
```

- "Based on the analysis of heart failure data, it is evident that the majority of fatalities occur within the age group of 55 to 75, with a median age of 65 years".

### Is there any significant difference in ejection fraction  between patients with and without diabetes?

```python
diabetes_mean = heart.groupby('diabetes')['ejection_fraction'].mean().sort_values()
print(diabetes_mean)

# Create a bar plot with ordered categories
plt.figure(figsize=(8, 6))
sns.barplot(x='diabetes', y='ejection_fraction', data=heart,order=diabetes_mean.index, errorbar=None)
plt.title('Mean Ejection Fraction by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Mean Ejection Fraction')
plt.show()
```

- The mean ejection fraction among patients with diabetes and those without it is approximately 38.
- This observation suggests that there is no substantial difference in the mean ejection fraction between patients with and without diabetes in the dataset.

### How does the serum creatinine level vary between patients who survived and those who did not? 

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=heart)
plt.title('Serum Creatinine Level by Survival Status')
plt.xlabel('Survival Status (0: Survived, 1: Deceased)')
plt.ylabel('Serum Creatinine Level')
plt.show()
```

```python
heart.groupby('DEATH_EVENT')['serum_creatinine'].agg(['mean','median','min','max'])
```

- This observation suggests a trend where patients who experienced a death event tend to have higher serum creatinine levels on average compared to those who survived. Additionally, the range of serum creatinine levels appears to be wider among deceased patients.

```python
X = heart.drop(['DEATH_EVENT'], axis=1)
corr = X.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, fmt='.2f')
plt.title('Correlation Among Independent Variables')
plt.show()
```

- Almost no correlation in between the independent features
